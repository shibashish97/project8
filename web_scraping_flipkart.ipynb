```python
url="https://www.flipkart.com/search?q=lenovo%20laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
import requests
from bs4 import BeautifulSoup
data=requests.get(url)
```

```python
data
```

```python
soup=BeautifulSoup(data.content)
```

```python
soup
```

```python
laptops=soup.find_all("div",class_="_4rR01T")
```

```python
laptops
```

```python
price=soup.find_all("div",class_="_30jeq3 _1_WHN1")
```

```python
price
```

```python
import pandas as pd
W=pd.DataFrame([laptops,price]).T
W.columns=["Laptop_Name","Price"]
```
